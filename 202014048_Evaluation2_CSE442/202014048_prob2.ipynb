```python
#Importing all the required libraries
import pandas as pd
import numpy as np
import sklearn
from pandas.core.dtypes.common import is_numeric_dtype
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
import seaborn as sns
import warnings as wr #Ignores the warnings
wr.filterwarnings('ignore')
df=pd.read_csv("https://raw.githubusercontent.com/shakil1819/CSE442-Machine-Learning-Sessional/main/202014048_Evaluation2_CSE442/Party%20Dataset.csv")
df
```

# Data Pre-processing

## Checking the NULL values

```python
df.drop('Party_Success', axis=1, inplace=True)
df.head()
```

### 1.7.5.2 Define dependent (target) and independent (predictor) features

```python
def fun(row):
    if row["Number_of_Guests"]<=10:
      if row["Party_Duration_Hours"]<=2.5:
        return "Party_Success:No"
      else:
        return "Party_Success:Yes"
    else:
      return "Party_Success:Yes"

df['Party_Success'] = df.apply(fun,axis=1)
df.head()
```

```python

y=df['Party_Success']
y
```

```python
x=df.drop('Party_Success', axis=1)
x
```

```python
clf = DecisionTreeClassifier()
clf.fit(x,y)
```

```python
clf.predict([[10,3,80]])
```

```python
text_rep = tree.export_text(clf)
```
